##### **Checking data integrity**

In this part we check if there are null values in the dataset.

```python
# Looking for NaN values in the dataset

nan_count=trains_df.isnull().sum()
print("NaN count in each column :")
nan_count
```

```python
#write afucntion that identify missing values in the dataset
def missing_values_table(df):
    #total missing values
    mis_val=df.isnull().sum()

    #percentage of missing values
    mis_val_percent=100*df.isnull().sum()/len(df)

    #make a table with the results
    mis_val_table=pd.concat([mis_val, mis_val_percent], axis=1)
    
    #rename the columns
    mis_val_table_ren_columns=mis_val_table.rename(columns={0:'Missing Values', 1:'% of Total Values'})

    #sort the table by percentage of missing descending
    mis_val_table_ren_columns=mis_val_table_ren_columns[mis_val_table_ren_columns.iloc[:,1]!=0].sort_values('% of Total Values', ascending=False).round(1)
    
    #return the dataframe with missing information
    return mis_val_table_ren_columns

#run the function to see the missing values
missing_values_table(trains_df)
```

```python
# Filter rows where 'Arrival Delay in Minutes' is NaN
missing_arrival_delay_rows = trains_df[trains_df['Arrival Delay in Minutes'].isnull()]

# Select columns to create distribution plots
selected_columns = ['Age', 'Gender', 'Loyalty', 'Work or Leisure', 'Ticket Class', 
                    'Onboard General Rating', 'Cleanliness Rating', 'Food\'n\'Drink Rating']

# Create a figure with subplots
fig, axes = plt.subplots(len(selected_columns), 1, figsize=(10, 15))

# Plot distribution plots for each selected column
for i, col in enumerate(selected_columns):
    sns.histplot(data=missing_arrival_delay_rows, x=col, ax=axes[i], kde=True, color='red', label='Missing Arrival Delay')
    sns.histplot(data=trains_df, x=col, ax=axes[i], kde=True, color='blue', label='Overall Dataset', alpha=0.5)
    axes[i].set_title(f'Distribution of {col} (Missing "Arrival Delay in Minutes")')
    axes[i].legend()

# Adjust spacing between subplots for better readability
plt.tight_layout()

# Show the plots
plt.show()
```

## **Data Processing**

---

##### **Partitioning the Dataset into Training and Test Set** 
In the following part we divide the dataset into train and test set. The train set is used to train the model,  while the testing segment is utilized to assess how well the model performs with data it hasn't encountered before.

This division is vital to mitigate the risk of overfitting, so to avoid that our model becomes excessively attuned to the training data, impairing its ability to adapt to new, unseen data effectively.
We used the common split ratio 80/20, where 80% of the data is allocated for training purposes and the remaining 20% is set aside for testing.

In this process, it's essential to remove the target variable from the training data. In our case, the column representing `Satisfied` is what we aim to predict.

```python
from sklearn.model_selection import train_test_split

# Splitting the dataset into features and target
x = trains.drop('Satisfied', axis=1)
y = trains['Satisfied'].iloc[:,0]

# Splitting the dataset into train and test sets
train, test, train_check, test_check = train_test_split(x, y, test_size=0.2, random_state=69)
```

---

##### **Scaling**

To prepare the data this process adjusts the range of different features to a standard scale, making sure that no single feature dominates the model due to its larger numeric range. In essence, scaling ensures that each feature contributes equally to the model's predictions, preventing bias towards variables simply because they have larger values.

```python
# Identify numerical and categorical columns 
numerical_columns_lst = x.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_columns_lst = x.select_dtypes(include=['object']).columns.tolist()
```

Here we set up a Pipeline that enables us to trasform categorical data in encoded numerical data, scales the data and returns the training and test set.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Create transformers for numerical and categorical data
numerical_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

# Combine transformers into a ColumnTransformer
transformation = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_columns_lst),
        ('cat', categorical_transformer, categorical_columns_lst)
    ])

# Create a pipeline with the ColumnTransformer
model_pipeline = Pipeline(steps=[('transformation', transformation)])

# Fit the pipeline to train and test set
train_encoded = model_pipeline.fit_transform(train)
test_encoded = model_pipeline.transform(test)
```

---

##### **Partitioning the training set into training and validation set**
Here we partition again the training set so that later we are able to fine-tune the model's hyperparameters and provide an unbiased evaluation of the model fit in the end.

The validation set is not used in the actual training process but it can be used to assess the model's performance at various stages of training.

```python
training, validation_test, training_check, validation_test_check = train_test_split(
    train_encoded, train_check, test_size=0.2, random_state=69)
```

## **Modelling**

---

We're dealing with a classification problem because our target variable is categorical, indicating whether customers are satisfied or not. To tackle this we're using four models:

- **Logistic Regression** 
    
    For binary classification problems with relatively simple, linear relationships between features and the target variable is a solid starting point to get information on the dataset, even if usually is not the most accurate model that is our primary goal.

- **K-Nearest Neighbors (KNN)** 
    
    Is an intuitive method that classifies a data point based on how its neighbors are classified. This model assumes that similar things exist in close proximity. In the context of customer satisfaction, it looks at the 'K' most similar customers (based on the features) to predict a customer's satisfaction. It's beneficial for capturing complex patterns without assuming a specific distribution in the data.

- **Decision Trees** 

    Offers a more flexible approach, handling complex patterns by dividing the data using a set of decision rules. This can be especially useful if there are interactions between features that affect customer satisfaction.

- **Random Forests** 

    build upon classification trees, using an ensemble of trees to improve predictive accuracy and control over-fitting. Given the likely complexity and potential noise in customer satisfaction data, the robustness of random forests makes them a strong candidate for improving prediction performance.

Selecting these four models allows us to compare a simple linear method with more complex non-linear methods, ensuring we explore both straightforward and intricate patterns in the data.

---

In the next phase of our analysis, we will rigorously evaluate each selected model (Logistic Regression, K-Nearest Neighbors (KNN), Decision Trees, and Random Forests) using five key metrics: accuracy, precision, recall, F1 score and AUC score. 
 
Through this metrics we'll be able to asses whether a model is suitable to our needs.

```python
# Importing score metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import confusion_matrix
```

---

##### **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

# Initializing the model
Logistic_Regression = LogisticRegression(max_iter=1000)

# Training the model
Logistic_Regression.fit(training, training_check)

# Predicting the target values from the validation data
prediction_LR = Logistic_Regression.predict(validation_test)

# Calculating the performance metrics
accuracy_LR = accuracy_score(validation_test_check, prediction_LR)
precision_LR = precision_score(validation_test_check, prediction_LR)
recall_LR = recall_score(validation_test_check, prediction_LR)
f1_LR = f1_score(validation_test_check, prediction_LR)

# Calculting the ROC Score
fpr, tpr, thresholds = roc_curve(validation_test_check, prediction_LR)
auc_LR = auc(fpr, tpr)

# Plotting the ROC Curve
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(fpr, tpr, marker='.')
plt.rcParams['figure.figsize'] = [4, 4]
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()

# Printing the results in a dataframe
results_lr = pd.DataFrame(data=[['Logistic Regression', accuracy_LR, precision_LR, recall_LR, f1_LR, auc_LR]], 
                        columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC'])
results_lr
```

```python
cm = confusion_matrix(validation_test_check, prediction_LR)
cm = pd.DataFrame(cm, index=['Actual Negative', 'Actual Positive'], columns=['Predicted Negative', 'Predicted Positive'])
cm
```

## **Hyperparameter Tuning**

---

After pre-testing with the validation test, we can now start finding the best parameters for the best models in our analisys (Random Forest, Decision Trees and KNN).

```python
# Importing the libraries
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

For each of the best models we are going to perform hyperparameter tuning by first using 'RandomizedSearchCV()' to get an idea of a small range of parameters around the ones that are found to be the best. Then by using 'GridSearchCV()' we input this previously found range and get parameters that improve the performance of the model.

Through this approach we ensure that we get a grid search that isn't extremely computationally expensive.

##### **Decision Tree**

Parameters tuned:
- `max_depth`: This parameter determines the maximum depth of the tree. A tree with a greater depth can model more complex relationships by adding more nodes, but it can also lead to overfitting. 

- `min_samples_split`: This parameter determines the minimum number of samples required to split an internal node. If the number of samples in a node is less than what we choose the node will not be split and will become a leaf node.

- `min_samples_leaf`: This parameter determines the minimum number of samples required to be at a leaf node. A split point at any depth will only be considered if it leaves at least the set number of training samples in each of the left and right branches.

```python
# Create the parameter grid
max_depth_dt = [int(x) for x in np.linspace(10, 40, num = 4)]
max_depth_dt.append(None)
min_samples_split_dt = [int(x) for x in np.linspace(2, 10, num = 5)]
min_samples_leaf_dt = [int(x) for x in np.linspace(1, 5, num = 5)]

# Create the random grid
random_grid_DT ={
    "max_depth": max_depth_dt,
    "min_samples_split":min_samples_split_dt, 
    "min_samples_leaf":min_samples_leaf_dt
}

# Create parameters
DT_randcv = RandomizedSearchCV(
    Decision_Tree, 
    random_grid_DT, 
    scoring="accuracy",
    n_jobs=-1, 
    verbose=1,
    n_iter=20,
    cv=5
    )

# Fit the random search model
DT_randcv.fit(training, training_check)

# Printing the best parameters
rand_bestparams_DT = DT_randcv.best_params_
print(f"Best paramters: {rand_bestparams_DT}")

# Initializing the model with the best parameters
DT_parameter = DecisionTreeClassifier(**rand_bestparams_DT, random_state=69)

# Training the model
DT_parameter.fit(training, training_check)

# Predicting the target values from the validation data
prediction_DT = DT_parameter.predict(validation_test)

# Calculating the performance metrics
accuracy_DT = accuracy_score(validation_test_check, prediction_DT)
precision_DT = precision_score(validation_test_check, prediction_DT)
recall_DT = recall_score(validation_test_check, prediction_DT)
f1_DT = f1_score(validation_test_check, prediction_DT)

# Calculting the ROC Score
fpr, tpr, thresholds = roc_curve(validation_test_check, prediction_DT)
auc_DT = auc(fpr, tpr)

# Plotting the ROC Curve
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(fpr, tpr, marker='.')
plt.rcParams['figure.figsize'] = [4, 4]
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()

# Printing the results in a dataframe
results_dt = pd.DataFrame(data=[['Decision Tree', accuracy_DT, precision_DT, recall_DT, f1_DT, auc_DT]], 
                        columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC'])
results_dt
```

##### **Random Forest**

Parameters tuned:
- `n_estimators`: This parameter determines the number of trees in the forest. More trees reduce the variance of predictions, improving the model's performance. However, adding more trees after a certain point does not improve the performance significantly, and it can be computationally expensive.

- `max_depth`: This parameter determines the maximum depth of the tree. A tree with a greater depth can model more complex relationships by adding more nodes, but it can also lead to overfitting. 

- `min_samples_split`: This parameter determines the minimum number of samples required to split an internal node. If the number of samples in a node is less than what we choose the node will not be split and will become a leaf node.

- `min_samples_leaf`: This parameter determines the minimum number of samples required to be at a leaf node. A split point at any depth will only be considered if it leaves at least the set number of training samples in each of the left and right branches.

```python
# Create a parameter grid 
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 250, num = 7)]
max_depth = list(range(1, 5))
max_depth.append(None)
min_samples_split =list(range(2, 4))
min_samples_leaf = list(range(1, 4))

# Create the random grid
random_grid_RF = {
    'n_estimators': n_estimators, 
    'max_depth': max_depth, 
    'min_samples_split': min_samples_split,
    'min_samples_leaf': min_samples_leaf, 
}

RF_randcv = RandomizedSearchCV(
    Random_Forest, 
    scoring='accuracy',
    param_distributions=random_grid_RF, 
    n_iter=30, 
    cv=5, 
    verbose=1, 
    random_state=69,
    n_jobs=-1
)

# Fit the random search model
RF_randcv.fit(training, training_check)

# Printing the best parameters
rand_bestparams_RF = RF_randcv.best_params_
print(f"Best paramters: {rand_bestparams_RF}")

# Initializing the model with the best parameters
RF_parameter = RandomForestClassifier(**rand_bestparams_RF,n_jobs=-1)

# Training the model
RF_parameter.fit(training, training_check)

#print all the scores
prediction_RF = RF_parameter.predict(validation_test)

# Calculating the performance metrics
accuracy_RF = accuracy_score(validation_test_check, prediction_RF)
precision_RF = precision_score(validation_test_check, prediction_RF)
recall_RF = recall_score(validation_test_check, prediction_RF)
f1_RF = f1_score(validation_test_check, prediction_RF)

# Calculting the ROC Score
fpr, tpr, thresholds = roc_curve(validation_test_check, prediction_RF)
auc_RF = auc(fpr, tpr)

# Plotting the ROC Curve
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(fpr, tpr, marker='.')
plt.rcParams['figure.figsize'] = [4, 4]
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()

# Printing the results in a dataframe
results_rf = pd.DataFrame(data=[['Random Forest', accuracy_RF, precision_RF, recall_RF, f1_RF, auc_RF]], 
                        columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC'])
results_rf
```

##### **KNN**

Parameters tuned:

- `n_neighbors`: This parameter determines the number of neighbors to use for majority voting or averaging in KNN. In other words, it decides the 'K' in KNN. A smaller value of 'K' makes the model more sensitive to noise, while a larger value makes it computationally expensive.

- `weights`: This parameter determines the weight function used in prediction. Possible values are 'uniform' (all points in each neighborhood are weighted equally), 'distance' (weight points by the inverse of their distance, closer neighbors of a query point will have a greater influence than neighbors which are further away), or any other user-defined function.

- `metric`: This parameter determines the distance metric used for the tree.The options include 'minkowski', 'manhattan', 'chebyshev', 'mahalanobis', etc. The choice of distance metric can greatly influence the performance of the model depending on the nature of the data.

```python
# Create a parameter grid
random_grid_KNN = {
    'n_neighbors': list(range(5, 15)),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

# Create the random grid
KNN_randcv = RandomizedSearchCV(
    KNN, 
    random_grid_KNN, 
    scoring="accuracy", 
    n_jobs=-1, 
    verbose=1, 
    n_iter=10,
    cv=5,
    )

# Fit the random search model
KNN_randcv.fit(training, training_check)

# Printing the best parameters
rand_bestparams_KNN = KNN_randcv.best_params_
print(f"Best paramters: {rand_bestparams_KNN})")

# Initializing the model with the best parameters
KNN_parameter = KNeighborsClassifier(**rand_bestparams_KNN, n_jobs=-1)

# Training the model
KNN_parameter.fit(training, training_check)

# Predicting the target values from the validation data
prediction_KNN = KNN_parameter.predict(validation_test)

# Calculating the performance metrics
accuracy_KNN = accuracy_score(validation_test_check, prediction_KNN)
precision_KNN = precision_score(validation_test_check, prediction_KNN)
recall_KNN = recall_score(validation_test_check, prediction_KNN)
f1_KNN = f1_score(validation_test_check, prediction_KNN)

# Calculting the ROC Score
fpr, tpr, thresholds = roc_curve(validation_test_check, prediction_KNN)
auc_KNN = auc(fpr, tpr)

# Plotting the ROC Curve
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(fpr, tpr, marker='.')
plt.rcParams['figure.figsize'] = [4, 4]
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()

# Printing the results in a dataframe
results_knn = pd.DataFrame(data=[['K Nearest Neighbors', accuracy_KNN, precision_KNN, recall_KNN, f1_KNN, auc_KNN]], 
                        columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC'])
results_knn
```
